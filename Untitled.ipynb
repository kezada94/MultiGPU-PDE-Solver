```python
from pyqtgraph.Qt import QtCore, QtGui
import pyqtgraph as pg
import pyqtgraph.opengl as gl
import numpy as np

## Create a GL View widgimport numpy as np
from mpl_toolkits import mplot3d
%matplotlib notebook

import matplotlib.pyplot as plt

import matplotlib.animation as anim
from IPython.display import HTML
from tqdm import tqdm
import time

import pyqtgraph as pg
import pyqtgraph.opengl as gl

Z = np.genfromtxt("result-32-32-10000.dat", delimiter=",")

Z = Z.reshape((-1, 32, 32))
app = QtGui.QApplication([])
w = gl.GLViewWidget()
#w.pan(3.1415, 3.1415, 0)
w.show()
w.setWindowTitle('pyqtgraph example: GLSurfacePlot')
w.setCameraPosition(distance=10)

## Add a grid to the view
g = gl.GLGridItem()

g.setDepthValue(10)  # draw grid after surfaces since they may be translucent
w.addItem(g)


## Animated example
## compute surface vertex data
W = 3.141592*2
H = 3.141592*2

cols = 32
rows = 32
L = 10000
x = np.linspace(0, W, cols).reshape(cols,1)
y = np.linspace(0, H, rows).reshape(1,rows)

cmap = plt.get_cmap('jet')

minZ=np.min(Z[0,:,:].reshape(-1))
maxZ=np.max(Z[0,:,:].reshape(-1))
rgba_img = cmap((Z[0,:,:].reshape(-1)-minZ)/(maxZ -minZ))

p4 = gl.GLSurfacePlotItem(x=x[:,0], y = y[0,:], shader='heightColor', colors=rgba_img, computeNormals=False, smooth=False)
#p4.scale(2,2,1)
p4.translate(-3.1415, -3.1415, 0)
p4.setData(z=Z[0,:,:])

w.addItem(p4)

index = 0
def update():
    global p4, z, index
    index += 1
    if(index >= L):
        index=0
    cmap = plt.get_cmap('jet')

    minZ=np.min(Z[index,:,:].reshape(-1))
    maxZ=np.max(Z[index,:,:].reshape(-1))
    rgba_img = cmap((Z[index,:,:].reshape(-1)-minZ)/(maxZ -minZ))
    p4.setColor(rgba_img)
    p4.setData(z=Z[index,:,:])
    
timer = QtCore.QTimer()
timer.timeout.connect(update)
timer.start(17)

## Start Qt event loop unless running in interactive mode.
if __name__ == '__main__':
    import sys
    if (sys.flags.interactive != 1) or not hasattr(QtCore, 'PYQT_VERSION'):
        QtGui.QApplication.instance().exec_()
```

```python
from pyqtgraph.Qt import QtGui, QtCore
import numpy as np
import pyqtgraph as pg
import matplotlib.pyplot as plt
from matplotlib import cm

import matplotlib.animation as anim
from IPython.display import HTML
from tqdm import tqdm
import time

import pyqtgraph as pg
import pyqtgraph.opengl as gl

Z = np.genfromtxt("result-6.283185-6.283185-64-64-1000.dat", delimiter=",")

Z = Z.reshape((-1, 64, 64))

app = QtGui.QApplication([])

## make pretty looping data
frames = 1000

data = Z



# Get the colormap
colormap = cm.get_cmap("magma")  # cm.get_cmap("CMRmap")
colormap._init()
lut = (colormap._lut * 255).view(np.ndarray)  # Convert matplotlib colormap from 0-1 to 0 -255 for Qt



win = pg.GraphicsLayoutWidget(show=True)
win.setWindowTitle('pyqtgraph example: Isocurve')
vb = win.addViewBox()
img = pg.ImageItem(data[0], lut=lut)
vb.addItem(img)
vb.setAspectLocked()

## generate empty curves
curves = []
levels = np.linspace(data.min(), data.max()*5)
for i in range(len(levels)):
    v = levels[i]
    ## generate isocurve with automatic color selection
    c = pg.IsocurveItem(level=v, pen=(i, len(levels)*1.5))
    c.setParentItem(img)  ## make sure isocurve is always correctly displayed over image
    c.setZValue(10)
    curves.append(c)

## animate!
ptr = 0
imgLevels = (data.min(), data.max() )
def update():
    global data, curves, img, ptr, imgLevels
    ptr = (ptr + 1) % data.shape[0]
    img.setImage(data[ptr], levels=imgLevels, lut=lut)
    for c in curves:
        c.setData(data[ptr])

timer = QtCore.QTimer()
timer.timeout.connect(update)
timer.start(50)
    
## Start Qt event loop unless running in interactive mode or using pyside.
if __name__ == '__main__':
    import sys
    if (sys.flags.interactive != 1) or not hasattr(QtCore, 'PYQT_VERSION'):
        QtGui.QApplication.instance().exec_()
```
