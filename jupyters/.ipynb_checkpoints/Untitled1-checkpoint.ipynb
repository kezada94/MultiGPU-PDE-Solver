```python
gauss = gauss*10
```

```python
tf = 10
L = 2

l1 = 10
l2 = 10
l3 = 10

dt = 0.001
dx = 0.01
dy = 0.01

M = int(L/dx)
N = int(L/dy)
T = int(tf/dt)

c = 0.5
C = 1/c*pow((1/dx)**2+(1/dy)**2, -1/2)

print(M, N, T, C)



u = np.zeros((T, M, N), dtype=np.float32)
v = np.zeros((T, M, N), dtype=np.float32)
w = np.zeros((T, M, N), dtype=np.float32)

t = np.linspace(0, tf, T)
x = np.linspace(0, L, M)
y = np.linspace(0, L, N)


u[0, int(M/2)-20:int(M/2)+20, N//2-20:N//2+20] = gauss
v[0, :40, :40] = gauss
w[0, -40:, -40:] = gauss

for i in tqdm(range(T)):
    if i == 0:
        continue
    elif i == 1:
        u[i, 1:-1, 1:-1] = 0.5*dt*dt*(-l1*np.sin(v[i-1, 1:-1, 1:-1] - w[i-1, 1:-1, 1:-1]) 
                          + ((-2*u[i-1, 1:-1, 1:-1]+u[i-1, 1:-1, :-2] + u[i-1, 1:-1, 2:])/(dy*dy)) 
                          + ((-2*u[i-1, 1:-1, 1:-1]+u[i-1, :-2, 1:-1] + u[i-1, 2:, 1:-1])/(dx*dx))) + u[i-1, 1:-1, 1:-1]
        v[i, 1:-1, 1:-1] = 0.5*dt*dt*(l2*np.sin(u[i-1, 1:-1, 1:-1] - w[i-1, 1:-1, 1:-1]) 
                          + ((-2*v[i-1, 1:-1, 1:-1]+v[i-1, 1:-1, :-2] + v[i-1, 1:-1, 2:])/(dy*dy)) 
                          + ((-2*v[i-1, 1:-1, 1:-1]+v[i-1, :-2, 1:-1] + v[i-1, 2:, 1:-1])/(dx*dx))) + v[i-1, 1:-1, 1:-1]
        w[i, 1:-1, 1:-1] = 0.5*dt*dt*(-l3*np.sin(u[i-1, 1:-1, 1:-1] - v[i-1, 1:-1, 1:-1]) 
                          + ((-2*w[i-1, 1:-1, 1:-1]+w[i-1, 1:-1, :-2] + w[i-1, 1:-1, 2:])/(dy*dy)) 
                          + ((-2*w[i-1, 1:-1, 1:-1]+w[i-1, :-2, 1:-1] + w[i-1, 2:, 1:-1])/(dx*dx))) + w[i-1, 1:-1, 1:-1]
    else:
        u[i, 2:-2, 2:-2] = dt*dt*(-l1*np.sin(v[i-1, 2:-2, 2:-2] - w[i-1, 2:-2, 2:-2]) 
                          + ((-2*u[i-1, 2:-2, 2:-2]+u[i-1, 2:-2, 1:-3] + u[i-1, 2:-2, 3:-1])/(dy*dy)) 
                          + ((-2*u[i-1, 2:-2, 2:-2]+u[i-1, 1:-3, 2:-2] + u[i-1, 3:-1, 2:-2])/(dx*dx))) - u[i-2, 2:-2, 2:-2]+ 2*u[i-1, 2:-2, 2:-2]
        
        v[i, 2:-2, 2:-2] = dt*dt*(l2*np.sin(u[i-1, 2:-2, 2:-2] - w[i-1, 2:-2, 2:-2]) 
                          + ((-2*v[i-1, 2:-2, 2:-2]+v[i-1, 2:-2, 1:-3] + v[i-1, 2:-2, 3:-1])/(dy*dy)) 
                          + ((-2*v[i-1, 2:-2, 2:-2]+v[i-1, 1:-3, 2:-2] + v[i-1, 3:-1, 2:-2])/(dx*dx))) - v[i-2, 2:-2, 2:-2]+ 2*v[i-1, 2:-2, 2:-2]
        
        w[i, 2:-2, 2:-2] = dt*dt*(-l1*np.sin(u[i-1, 2:-2, 2:-2] - v[i-1, 2:-2, 2:-2]) 
                          + ((-2*w[i-1, 2:-2, 2:-2]+w[i-1, 2:-2, 1:-3] + w[i-1, 2:-2, 3:-1])/(dy*dy)) 
                          + ((-2*w[i-1, 2:-2, 2:-2]+w[i-1, 1:-3, 2:-2] + w[i-1, 3:-1, 2:-2])/(dx*dx))) - w[i-2, 2:-2, 2:-2]+ 2*w[i-1, 2:-2, 2:-2]
        
```

```python
from pyqtgraph.Qt import QtCore, QtGui
import pyqtgraph as pg
import pyqtgraph.opengl as gl
import numpy as np

## Create a GL View widgimport numpy as np
from mpl_toolkits import mplot3d
%matplotlib notebook

import matplotlib.pyplot as plt

import matplotlib.animation as anim
from IPython.display import HTML
from tqdm import tqdm
import time

import pyqtgraph as pg
import pyqtgraph.opengl as gl


Z = u
Z2 = v
Z3 = w
app = QtGui.QApplication([])
w = gl.GLViewWidget()
#w.pan(3.1415, 3.1415, 0)
w.setWindowTitle('pyqtgraph example: GLSurfacePlot')
w.setCameraPosition(distance=10)

## Add a grid to the view
g = gl.GLGridItem()

g.setDepthValue(10)  # draw grid after surfaces since they may be translucent
w.addItem(g)
w.show()
## Animated example
## compute surface vertex data


cols = len(x)
rows = len(y)
L = 10000
xx = x.reshape(cols,1)
yy = y.reshape(1,rows)

cmap = plt.get_cmap('jet')
cmap2 = plt.get_cmap('jet')
cmap3 = plt.get_cmap('jet')

minZ=np.min(Z[0,:,:].reshape(-1))
maxZ=np.max(Z[0,:,:].reshape(-1))
rgba_img = cmap((Z[0,:,:].reshape(-1)-minZ)/(maxZ -minZ))

minZ=np.min(Z2[0,:,:].reshape(-1))
maxZ=np.max(Z2[0,:,:].reshape(-1))
rgba_img2 = cmap2((Z2[0,:,:].reshape(-1)-minZ)/(maxZ -minZ))

minZ=np.min(Z3[0,:,:].reshape(-1))
maxZ=np.max(Z3[0,:,:].reshape(-1))
rgba_img3 = cmap3((Z3[0,:,:].reshape(-1)-minZ)/(maxZ -minZ))

p4 = gl.GLSurfacePlotItem(x=xx[:,0], y = yy[0,:], colors=rgba_img, computeNormals=False, smooth=False)
p5 = gl.GLSurfacePlotItem(x=xx[:,0], y = yy[0,:], colors=rgba_img2, computeNormals=False, smooth=False)
p6 = gl.GLSurfacePlotItem(x=xx[:,0], y = yy[0,:], colors=rgba_img3, computeNormals=False, smooth=False)
#p4.scale(2,2,1)
p4.translate(-1, -1, 0)
p4.setData(z=Z[0,:,:])

p5.translate(-1, -1, 1)
p5.setData(z=Z2[0,:,:])

p6.translate(-1, -1, 2)
p6.setData(z=Z3[0,:,:])

w.addItem(p4)
w.addItem(p5)
w.addItem(p6)

index = 0
def update():
    global p4, p5, p6, Z, Z2, Z3, index
    index += 1
    if(index >= L):
        index=0
    cmap = plt.get_cmap('jet')
    cmap2 = plt.get_cmap('jet')
    cmap3 = plt.get_cmap('jet')
    
    
    minZ=np.min(Z[index,:,:].reshape(-1))
    maxZ=np.max(Z[index,:,:].reshape(-1))
    rgba_img = cmap((Z[index,:,:].reshape(-1)-minZ)/(maxZ -minZ))
    p4.setData(z=Z[index,:,:], colors=rgba_img)

    minZ=np.min(Z2[index,:,:].reshape(-1))
    maxZ=np.max(Z2[index,:,:].reshape(-1))
    rgba_img2 = cmap2((Z2[index,:,:].reshape(-1)-minZ)/(maxZ -minZ))
    p5.setData(z=Z2[index,:,:], colors=rgba_img2)

    minZ=np.min(Z3[index,:,:].reshape(-1))
    maxZ=np.max(Z3[index,:,:].reshape(-1))
    rgba_img3 = cmap3((Z3[index,:,:].reshape(-1)-minZ)/(maxZ -minZ))
    p6.setData(z=Z3[index,:,:], colors=rgba_img3)

    
timer = QtCore.QTimer()
timer.timeout.connect(update)
timer.start(17)

## Start Qt event loop unless running in interactive mode.
if __name__ == '__main__':
    import sys
    #if (sys.flags.interactive != 1) or not hasattr(QtCore, 'PYQT_VERSION'):
    QtGui.QApplication.instance().exec_()
```
